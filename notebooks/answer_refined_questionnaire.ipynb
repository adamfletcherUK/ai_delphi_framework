```python
import glob

personas = glob.glob('outputs/6_final_personas/*.md')
questionnaires = glob.glob('outputs/11_further_questions/*.md')

# with open('outputs/11_further_questions/How do you anticipate changes in regulatory requirements with the transition to a modern infrastructure, and how do you plan to navigate them effectively?.md') as q:
#     questionnaire = q.read()

for persona in personas:
    with open(persona) as p:
        persona_content = p.read()

    print(os.path.basename(persona).replace('.md',''))

    chat_template = ChatPromptTemplate.from_messages(
        [
            SystemMessage(
                content=(f"""{persona_content}
                        
                        {open('./persona_descriptions/answer_refined_questionnaires.txt').read()}
                        """)
            ),
            HumanMessagePromptTemplate.from_template(
                """Can you to the best of your ability answer this questionnaire giving very detailed responses: {questionnaire_content}"""
                ),
        ]
    )

    llm = ChatOpenAI(model='gpt-4-turbo-preview') ## make this defined ones and used again and again!!

    # answered_questionnaire = llm.invoke(chat_template.format_messages(questionnaire=questionnaire)).content

    # with open(f"outputs/12_answered_refined_questionnaires/{os.path.basename(persona).replace('.md','')}.md", "a") as text_file:
    #     text_file.write(f"""{answered_questionnaire}
        
    #                     """)

    for questionnaire in questionnaires:
        with open(questionnaire) as q:
            questionnaire_content = q.read()
        print(f"  - {os.path.basename(questionnaire).replace('.md','')}")

        answered_questionnaire = llm.invoke(chat_template.format_messages(questionnaire_content=questionnaire_content)).content

        with open(f"outputs/12_answered_refined_questionnaires/{os.path.basename(persona).replace('.md','')}.md", "a") as text_file:
            text_file.write(f"""{answered_questionnaire}
                        
""")
```
